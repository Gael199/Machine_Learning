# DECISION TREE

### Importing the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
music_data = pd.read_csv("music.csv")
music_data.head()
```

### Defining the features and the target

```python
X = music_data.iloc[:,0:2]
y = music_data.iloc[:,-1]
```

### Splitting the the Data into the Training set and Test set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.2, random_state = 0)
```

### Training the Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

### Predicting the Test results 

```python
y_pred = classifier.predict(X_test)
```

### Evaluating and improve

### Making the confusion Matrix

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,y_pred)
print(cm)
```

### Calculationg the accuracy 

```python
from sklearn.metrics import accuracy_score

score = accuracy_score(y_test, y_pred)
score
```

### Visualization

```python
from sklearn import tree

tree.plot_tree(classifier, feature_names = ['age','gender'], class_names = sorted(y.unique()), filled = True)
plt.show()
```
